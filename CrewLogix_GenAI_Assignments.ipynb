# **Creating an CLI-based ChatBot using google gemini-api**

```python
from google import genai
from google.colab import userdata
import json
```

```python
from google.colab import userdata
api_key = userdata.get('GEMINI_API_KEY')
```

```python
messages=[]
```

```python
def agent_response(prompt):
  client = genai.Client(api_key = api_key)

  response = client.models.generate_content_stream(
    model="gemini-3-flash-preview",
    contents=f"{prompt}",
  )
  full_text = ""
  for chunk in response:
        if chunk.text:
            print(chunk.text, end="", flush=True) # Print immediately
            full_text += chunk.text

  messages.append({"user_prompt":prompt,"agent_response":full_text})
  print("\n")

```

```python
def display():
  print(json.dumps(messages,indent=4))
```

```python
if __name__ == "__main__":
  user_prompt = input("Ask Anything! ")

  while user_prompt != "quit":
    agent_response(user_prompt)
    user_prompt = input("Ask Anything! ")
```

```python
display()
```
